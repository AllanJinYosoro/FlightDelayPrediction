```python
import sys
sys.path.append('..')

from model.ARIMA import pipline

import pandas as pd
```

```python
data = pd.read_csv('../data\scheduled_flight\continuous_flights_data.csv')
```

```python
flights=[]
fl=['AA1042','AA1406','AA1486','AA1567','AA167']
for i in fl:
    flights.append(data.loc[data['FLIGHT_ID'] == i])
```

```python
flights[0].columns
```

```python
P1 = pipline(flights[0],False,True)
P2 = pipline(flights[1],False,True)
P3 = pipline(flights[2],False,True)
P4 = pipline(flights[3],False,True)
P5 = pipline(flights[4],False,True)

piplines = [P1,P2,P3,P4,P5]
```

```python
for p in piplines:
    p.fit(False)
    p.summary()
```
